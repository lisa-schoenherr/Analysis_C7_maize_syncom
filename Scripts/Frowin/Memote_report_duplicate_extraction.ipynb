## Imports

```python
import webbrowser
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import ast
import os
import json
import time
```

## Paths

```python
file_path = '/home/frowin/Desktop/master_thesis/model_preparation/root_reports_all_modified/'
```

## Test file

```python
test_file = ['modified_Root4_report.html']
```

The list of mass imbalanced reactions can not be accessed in the same way as scores. Since the list is behind a dropdown menu the html file must not only be opened  using a webdriver but additionally the corrsponding dropdown menu must be openend by 'clicking' on the header 'Mass Balance'.  
The document can then be searched for the type of html element that contains the list (as a string) and the list can be extracted.

```python
# Create a dict to store all results
memote_duplicates = {}

# Iterate through all MEMOTE reports in the dict
for file in test_file:

    file_name = file[:-12]

    memote_duplicates[file_name] = {}
        
    # Initialize WebDriver
    driver = webdriver.Chrome()

    # Open the HTML file 
    driver.get(f"file://{file_path}{file}")

    # Wait for the dropdown button for the metabolite duplicates test to load and click it
    dropdown_button1 = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "mat-expansion-panel-header-106"))
    )
    dropdown_button1.click()

    time.sleep(2)

    # Wait for the dropdown button for the Reactions With Partially Identical Annotations to load and click it
    dropdown_button2 = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "mat-expansion-panel-header-114"))
    )    

    dropdown_button2.click()

    time.sleep(2)

    # Wait for the dropdown button for the Duplicate Reactions to load and click it
    dropdown_button3 = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "mat-expansion-panel-header-115"))
    )

    dropdown_button3.click()

    time.sleep(2)

    # Wait for the dropdown button for the Reactions with identical genes to load and click it
    dropdown_button4 = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, "mat-expansion-panel-header-116"))
    )

    dropdown_button4.click()

    time.sleep(2)

    # Wait for all dropdown items to appear and assign all elements with the type CSS_SELECTOR = [ng-reflect-ng-switch] as a list
    # These elements can also be assessed without opening the dropdown menu but the list of reactions won't be found then
    elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, '[ng-reflect-ng-switch]'))
    )

    '''
    # This step doesn't has any function but the list of elements is not accessible without printing it once first?!
    print(elements[93].text)
    print(elements[102].text)
    print(elements[104].text)
    print(elements[106].text)
    '''

    # Extract the string containing the list of reactions
    metabolite_duplicates_string = elements[93].text.strip()
    annotation_duplicates_string = elements[102].text.strip()
    reaction_duplicates_string = elements[104].text.strip()
    gpr_duplicates_string = elements[106].text.strip()


    # FORMAT AND SAVE METABOLIC DUPLICATES

    # Transform the string into a list
    metabolite_duplicates_list = ast.literal_eval(metabolite_duplicates_string)

    # Put the list into a dictionary with the model id as key
    memote_duplicates[file_name]['Metabolite Duplicates'] = metabolite_duplicates_list


    # FORMAT AND SAVE ANNOTATION DUPLICATES

    # Transform the string into a list
    annotation_duplicates_list = ast.literal_eval(annotation_duplicates_string)

    # Put the list into a dictionary with the model id as key
    memote_duplicates[file_name]['Annotation Duplicates'] = annotation_duplicates_list


    # FORMAT AND SAVE REACTION DUPLICATES

    # Transform the string into a list
    reaction_duplicates_list = ast.literal_eval(reaction_duplicates_string)

    # Put the list into a dictionary with the model id as key
    memote_duplicates[file_name]['Reaction Duplicates'] = reaction_duplicates_list


    # FORMAT AND SAVE GPR DUPLICATES

    # Transform the string into a list
    gpr_duplicates = ast.literal_eval(gpr_duplicates_string)

    # Put the list into a dictionary with the model id as key
    memote_duplicates[file_name]['GPR Duplicates'] = gpr_duplicates

    # Quit the driver
    driver.quit()

# Save the results of all iterations as json 
with open('Duplicates_in_all_models_memote.json', 'w') as file:
    json.dump(memote_duplicates, file, indent=4)

    
```
